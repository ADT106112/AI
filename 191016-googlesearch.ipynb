```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.google.com.tw/search'
my_keyword = {'q':'玉龍'}
r = requests.get(url, my_keyword)

if r.status_code == requests.codes.ok:
    soup = BeautifulSoup(r.text, 'html.parser')
    
    print(soup.prettify())
```
